```python
# Creates a schema from a known good month

sdf_jan2024 = spark.read.parquet('../data/tlc_data/landing/2024-01.parquet')
consistent_col_casing = [F.col(col_name).alias(col_name.lower()) for col_name in sdf_jan2024.columns]
sdf_jan2024 = sdf_jan2024.select(*consistent_col_casing)

schema = sdf_jan2024.schema

with open("../scripts/schema.json", "w") as f:
    json.dump(schema.jsonValue(), f)
```

```python
with open('../scripts/schema.json', 'r') as file:  
    schema = StructType.fromJson(json.load(file))
```

```python
# Manually cast any bad parquet files.

sdf_malformed = spark.read.parquet('../data/tlc_data/landing/2023-01.parquet')

# select all columns from the existing malformed dataframe and cast it to the required schema
sdf_malformed = sdf_malformed \
    .select([F.col(c).cast(schema[i].dataType) for i, c in enumerate(sdf_malformed.columns)])
    
sdf_malformed \
    .coalesce(1) \
    .write \
    .mode('overwrite') \
    .parquet('../data/tlc_data/raw/2023-01.parquet')
```

```python
# Fails on stage 7 sometimes, just rerun the script. Not sure what the issue is.
# Ensures that the schema is consistent across all months

for year in [2023, 2024]:
    for month in range(1, 7):
        if month == 1 and year == 2023:
            continue
        input_path = f'../data/tlc_data/landing/{str(year)}-{str(month).zfill(2)}.parquet'
        output_path = f'../data/tlc_data/raw/{str(year)}-{str(month).zfill(2)}.parquet'
        
        # read and write in one single statement, not great but it is what it is
        # do not copy this, this is not best practice
        sdf = spark \
            .read \
            .schema(schema) \
            .parquet(input_path) \
                
        sdf \
            .coalesce(1) \
            .write \
            .mode('overwrite') \
            .parquet(output_path)
```

```python
# Read in weather data
year1_df = pd.read_csv('../data/weather_data/landing/2023_weather.csv')
year2_df = pd.read_csv('../data/weather_data/landing/2024_weather.csv')
```

```python
# Concatenate the two years, drop columns that are not needed and/or constantly null
weather_df = pd.concat([year1_df, year2_df], ignore_index = False)
weather_df.drop(columns=['wdir', 'wpgt', 'snow','tsun', 'coco'], inplace=True)
```

```python
# Convert the time column to datetime, extract the date and hour, and drop the time column
weather_df['time'] = pd.to_datetime(weather_df['time'])
weather_df['date'] = weather_df['time'].dt.date
weather_df['hour'] = weather_df['time'].dt.hour
weather_df.drop(columns=['time'], inplace=True)
```

```python
# Rearrange columns to have 'date' and 'hour' first
weather_df = weather_df[['date', 'hour'] + [col for col in weather_df.columns if col not in ['date', 'hour']]]

weather_df.to_csv('../data/weather_data/curated/New_York.csv', header=True, index=False)
```
